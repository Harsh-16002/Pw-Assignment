```python
'''
5. What are iterators in Python and how do they differ from iterables?

 In Python, iterators and iterables are both used to represent sequences of elements, but they have distinct characteristics and functionalities.
 Here's a breakdown of what they are and how they differ:

 1. Iterable
 An iterable is any Python object that can be looped over or iterated through. These objects have an internal method __iter__() that returns an iterator. Common examples of iterables include:

 Lists: [1, 2, 3]
 Tuples: (1, 2, 3)
 Strings: "hello"
 Dictionaries: {"key": "value"}
 Sets: {1, 2, 3}
 Characteristics of an Iterable:
 Can be iterated: You can use a for loop to iterate over its elements.
 Has __iter__() method: An iterable object implements this method, which allows it to return an iterator.
 Example:  '''

my_list = [1, 2, 3]

for element in my_list:  # my_list is an iterable
    print(element)
print("From Iterable")



 # In this example, my_list is an iterable because we can use a for loop to go through its elements.

'''
 2. Iterator
 An iterator is an object that represents a stream of data and can return the next item in the sequence on demand. An iterator object must implement two methods:

 __iter__() - returns the iterator object itself.
 __next__() - returns the next element in the sequence, and raises a StopIteration exception when no more items are available.
 Characteristics of an Iterator:
 Consumes items: Once an item is iterated over, it is consumed and cannot be re-accessed unless you create the iterator again.
 Lazy evaluation: It produces values on demand, which means it doesn't store all the values in memory. This is useful when dealing with large data sets.
 Has __next__() method: The next() function can be called to get the next item in the iterator.
 Example:'''

my_list = [1, 2, 3]
my_iterator = iter(my_list)  # Create an iterator from the iterable

try:
    print(next(my_iterator))  # Output: 1
    print(next(my_iterator))  # Output: 2
    print(next(my_iterator))  # Output: 3
    print(next(my_iterator))  # Raises StopIteration
except StopIteration:
    print("Iterator is exhausted.")




# Key differences:

# Iterables are the objects being iterated over, while iterators are the objects doing the iterating.
# Iterables can be reused, while iterators are exhausted after a single use.



 # In this example, my_iterator is an iterator that we created from the iterable my_list.
 # The next() function is used to fetch the next element from the iterator until the sequence is exhausted.

```

```python
'''
6. Explain the concept of generators in Python and how they are defined.

Generators in Python are a special type of iterator that allow you to iterate over a sequence of values lazily, meaning they produce values on the fly rather than storing them in memory.
This is especially useful when dealing with large datasets or streams of data that don't fit into memory all at once.

Key Concepts:

Lazy Evaluation: Generators compute and return values only when requested. This makes them memory efficient.

Yield Statement: Generators use the yield keyword to return values one at a time. Unlike return, which ends the function execution, yield pauses the function and preserves its state for the next iteration.

Iterator Protocol: Generators automatically implement the iterator protocol, meaning they have __iter__() and __next__() methods.

Defining a Generator
A generator is defined like a normal function, but instead of using the return keyword to return a value, it uses the yield keyword.
Each time yield is encountered, the function’s state is saved, and the yielded value is returned.
When next() is called again, the function resumes execution just after the last yield.

Example of a Generator
'''
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Using the generator
counter = count_up_to(3)

print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3


```

```python
'''
 7. What are the advantages of using generators over regular functions?

Generators have several advantages over regular functions:

1. Memory Efficiency: Generators use significantly less memory than regular functions because they only store the current state of the iteration, not the entire dataset.

2. Lazy Evaluation: Generators only compute values on demand, which can be more efficient than computing all values at once.

3. Flexibility: Generators can be used to create complex data pipelines or workflows.

4. Improved Performance: Generators can perform better than regular functions for large datasets.

5. Simplified Code: Generators can simplify code by eliminating the need for unnecessary loops or conditional statements.

6. Easier Debugging: Generators can make debugging easier since they allow for incremental execution.

7. Better Support for Infinite Data Structures: Generators can handle infinite data structures, like infinite lists or streams, more naturally than regular functions.

8. Improved Readability: Generators can improve code readability by separating the iteration logic from the computation logic.

9. Easier Parallelization: Generators can make parallelizing code easier since they can be used to create independent, concurrent tasks.

10. Better Support for Cooperative Multitasking: Generators can be used to implement cooperative multitasking, where tasks voluntarily yield control to other tasks.

Overall, generators provide a powerful and flexible way to handle data iteration and computation, making them a valuable tool in Python programming.

Here is an example of a generator function in Python:
'''

def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()

for value in gen:
    print(value)

# Output:
# 1
# 2
# 3


# In this example, my_generator is a generator function that yields three values: 1, 2, and 3. The yield keyword is used to define a generator function, and it specifies the values that the generator will produce.

# When we create a generator object gen by calling my_generator(), we can iterate over the generator using a for loop. The generator will produce the values 1, 2, and 3, one at a time, and the for loop will print each value.
```

```python
'''
 9. Explain the purpose and usage of the `map()` function in Python.

 The map() function in Python is a built-in function that applies a given function to each item of an iterable (such as a list, tuple, or dictionary) and returns a list of the results.
 The purpose of map() is to process and transform data in a concise and readable way.

Syntax:

map(function, iterable)

Where:

 function is the function to be applied to each item of the iterable
 iterable is the iterable to be processed

Usage:

 Apply a function to each item of a list:
'''
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)  # [1, 4, 9, 16, 25]

# Convert a list of strings to uppercase:

words = ["hello", "world"]
uppercase_words = list(map(str.upper, words))
print(uppercase_words)  # ["HELLO", "WORLD"]

# Use map() with multiple iterables:

numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
summed_numbers = list(map(lambda x, y: x + y, numbers1, numbers2))
print(summed_numbers)  # [5, 7, 9]
```

```python
'''
 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 The functions map(), reduce(), and filter() in Python are used to apply a function to sequences like lists, tuples, etc. However, they each serve different purposes. Here's how they differ:

1. map():

Purpose: Applies a given function to all items in an iterable (like a list) and returns a map object (an iterator) with the results.
Function Signature: map(function, iterable)
Behavior: It transforms each element of the iterable based on the function.
Use case: When you need to apply the same operation to each item in a sequence.
Example:

'''
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]

'''

2. filter():
Purpose: Filters the elements in an iterable based on a condition defined by a function. Only the elements for which the function returns True are kept.
Function Signature: filter(function, iterable)
Behavior: It filters out elements from the iterable that don't meet a certain condition.
Use case: When you need to remove elements from a sequence that don’t satisfy a condition.

Example
'''
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4, 6]

'''
3. reduce():
Purpose: Applies a function cumulatively to the items of an iterable, reducing the iterable to a single value.
Function Signature: reduce(function, iterable[, initializer])
Behavior: It combines all elements into a single value by applying the function between pairs of elements.
Use case: When you need to perform a rolling computation to reduce the sequence to a single result. For example, summing or multiplying all elements in a list.

Example
'''
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24 (1*2*3*4)


# Key Differences:
# map() transforms every element in the iterable by applying a function to them.
# filter() extracts elements that meet a certain condition defined by the function.
# reduce() combines all elements into a single result using a cumulative function, often used for things like summing or multiplying all the elements in a list.
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200


def is_prime_in_range(n):

    if n <= 1 or n > 200:
        return False

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

number = 157
if is_prime_in_range(number):
    print(f"{number} is a prime number and it is within the range of 1 to 200.")
else:
    print(f"{number} is either not a prime number or it is outside the range of 1 to 200.")

```

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current_term = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.num_terms:
            raise StopIteration
        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:

            next_fib = self.a + self.b
            self.a, self.b = self.b, next_fib
            self.current_term += 1
            return next_fib


num_terms = 10
fib_iterator = FibonacciIterator(num_terms)

print(f"Fibonacci sequence up to {num_terms} terms:")
for fib_num in fib_iterator:
    print(fib_num)
```

```python
 #6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

 def powers_of_2(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

max_exp = 5
for power in powers_of_2(max_exp):
    print(power)
```

```python
#  7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()


file_path = 'Sample.txt'
for line in read_file_line_by_line(file_path):
    print(line)
```
