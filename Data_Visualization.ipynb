```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the style for seaborn
sns.set(style="whitegrid")

# Generating a synthetic dataset
np.random.seed(42)  # For reproducibility
n_samples = 100

# Creating correlated features
x = np.random.normal(0, 1, n_samples)
y = 2 * x + np.random.normal(0, 0.5, n_samples)  # y is correlated with x
z = 0.5 * x + 0.5 * y + np.random.normal(0, 0.5, n_samples)  # z is correlated with x and y

# Creating a DataFrame
data = pd.DataFrame({
    'Feature 1': x,
    'Feature 2': y,
    'Feature 3': z
})

# Calculating the correlation matrix
corr_matrix = data.corr()

# Creating a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar=True)

# Adding title
plt.title("Correlation Matrix Heatmap")

# Display the plot
plt.show()
```

```python
'''
5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK',
'Germany', 'France'],
    'Population':
np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000,
5)
}
df = pd.DataFrame(data)
'''

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set the random seed for reproducibility
np.random.seed(25)

# Create the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create the bubble chart
plt.figure(figsize=(12, 8))
plt.scatter(df['GDP'], df['Population'],
            s=df['Population'] * 10,  # Bubble size proportional to population
            alpha=0.6,
            c='blue',
            edgecolors='w',
            linewidth=2)

# Add titles and labels
plt.title('Bubble Chart of Country Population vs GDP', fontsize=16)
plt.xlabel('GDP (in million USD)', fontsize=14)
plt.ylabel('Population', fontsize=14)

# Add grid
plt.grid(True)

# Add annotations for each country
for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['GDP'][i], df['Population'][i]), fontsize=10, ha='right')

# Show the plot
plt.show()
```

```python
'''
3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
'''

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Convert data to a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Set up the output
output_notebook()  # Use this if you're running in a Jupyter Notebook
# For script execution, use: output_file("bar_chart.html")

# Create a new plot
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts',
           width=800, height=400)

# Add a bar renderer
p.vbar(x='fruits', top='counts', source=source, width=0.4,
        color='blue', legend_label="Counts")

# Customize the legend and grid
p.legend.location = "top_left"
p.grid.grid_line_color = "gray"

# Show the plot
show(p)


```

```python
'''
4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
'''
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data
data_hist = np.random.randn(1000)  # 1000 random numbers from a standard normal distribution
hist, edges = np.histogram(data_hist, bins=30)  # Calculate histogram

# Set up the output
output_notebook()  # Use this if you're running in a Jupyter Notebook
# For script execution, use: output_file("histogram.html")

# Create a new plot
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           width=800, height=400)

# Add a quad renderer to create the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color='blue', line_color='black', alpha=0.7)

# Show the plot
show(p)

```

```python
'''
5. Create a Bokeh heatmap using the provided dataset

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
'''
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.palettes import Inferno256  # Alternative color palette

# Generate random data
data_heatmap = np.random.rand(10, 10)  # Random values for the heatmap
x = np.linspace(0, 1, 10)  # x-coordinates
y = np.linspace(0, 1, 10)  # y-coordinates

# Create a meshgrid for x and y coordinates
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Set up the output
output_notebook()  # Use this if you're running in a Jupyter Notebook
# For script execution, use: output_file("heatmap.html")

# Create a new plot
p = figure(title="Heatmap Example",
           x_axis_label='X',
           y_axis_label='Y',
           width=600, height=600)

# Create a color mapper
mapper = linear_cmap(field_name='z', palette=Inferno256, low=min(z_flat), high=max(z_flat))

# Add rectangles to create the heatmap
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, z=z_flat))
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        fill_color=mapper, line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)

```
